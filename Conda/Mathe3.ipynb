**Erzeugen Sie dann einen Vektor t mit den Zahlen [1, 1.01, 1.0,..., 2]**
numpy Arrays können als mehrdimensionale Vektoren behandelt werden, daher hier die ausgabe als Array.
Um den Vektor nicht aufwendig per Hand implementieren zu müssen greife ich auf eine Funktion zurück, die es mir erlaubt ein intervall und die schrittbreite anzugeben und die daraus resultierenden Zahlen als eindimensionales `np.array` auszugeben.

$x \in \mathbb{R}^1$

```python
t = np.arange(1,2.01,0.01)
print("t = ",t)
```

um jedes Elemnt aus t zu Quadrieren schreibe ich eine kleine Funktion um den Code übersichtlich zu halten.

`np.vectorize(<lambda>)` nimmt eine Funktion als Parameter und wendet diese Funktion auf jedes Elment des `np.array` an.

```python
square = np.vectorize(lambda x: x**2)

y = square(t)
print("t² = ", y)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(t,y, label='$y = t^2$, $x = t$')
#plt.title('Neilsche Parabel')
plt.xlabel('$x = t$')
plt.ylabel('$y = t^2$')
plt.grid(True)
plt.legend()
plt.show()
```

**Anwendung** plotte die Neilsche Parabel $f(t) = (t^2, t^3)$

```python
#negative Werte zeigen die Parabel besser
t = np.arange(-2,2,0.01)

t_pow3 = np.vectorize(lambda x: x ** 3)(t)
t_pow2 = np.vectorize(lambda x: x ** 2)(t)

plt.figure(figsize=(8, 6))
plt.plot(t_pow2,t_pow3, label='Neilsche Parabel $y = t^3$, $x = t^2$')
plt.title('Neilsche Parabel')
plt.xlabel('$x = t^2$')
plt.ylabel('$y = t^3$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
A = np.array([[2,1,3],
              [2,3,4],
              [5,3,7]])

# Hauptdiagonale mit 2en der Größe 10
main_diag = 2 * np.ones(10)

# Nebendiagonalen mit 1en der Größe 9
off_diag = 1 * np.ones(9)

# Erstellen der Matrix T
T = np.diag(main_diag) - np.diag(off_diag, 1) - np.diag(off_diag, -1)

print("Matrix A:")
print(A)
print("\n")

print("Matrix T:")
print(T)
```

Sind matrix $T$ und $A$ als lineares Gleichungssystem lösbar?

dafür müssen wir die Determinanten bestimmen, sind diese nicht $0$, dann ist das Gleichungssystem mit der gegebenen Matrix eindeutig lösbar.

```python
det_T = np.linalg.det(T)
print(f"Determinate von Matrix T: {np.round(det_T,decimals=3)}")
print ("T ist eindeutig bestimmbar" if det_T != 0 else "T ist nicht einduetig bestimmbar")

print("\n")
det_A = np.linalg.det(A)
print(f"Determinate von Matrix A: {np.round(det_A,decimals=3)}")
print ("A ist eindeutig bestimmbar" if det_A != 0 else "A ist nicht einduetig bestimmbar")
```
