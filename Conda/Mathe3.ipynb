**Erzeugen Sie dann einen Vektor t mit den Zahlen [1, 1.01, 1.0,..., 2]**
numpy Arrays können als mehrdimensionale Vektoren behandelt werden, daher hier die ausgabe als Array.
Um den Vektor nicht aufwendig per Hand implementieren zu müssen greife ich auf eine Funktion zurück, die es mir erlaubt ein intervall und die schrittbreite anzugeben und die daraus resultierenden Zahlen als eindimensionales `np.array` auszugeben.

$x \in \mathbb{R}^1$

```python
t = np.arange(1,2.01,0.01)
print("t = ",t)
```

um jedes Elemnt aus t zu Quadrieren schreibe ich eine kleine Funktion um den Code übersichtlich zu halten.

`np.vectorize(<lambda>)` nimmt eine Funktion als Parameter und wendet diese Funktion auf jedes Elment des `np.array` an.

```python
square = np.vectorize(lambda x: x**2)

y = square(t)
print("t² = ", y)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(t,y, label='$y = t^2$, $x = t$')
#plt.title('Neilsche Parabel')
plt.xlabel('$x = t$')
plt.ylabel('$y = t^2$')
plt.grid(True)
plt.legend()
plt.show()
```

**Anwendung** plotte die Neilsche Parabel $f(t) = (t^2, t^3)$

```python
#negative Werte zeigen die Parabel besser
t = np.arange(-2,2,0.01)

t_pow3 = np.vectorize(lambda x: x ** 3)(t)
t_pow2 = np.vectorize(lambda x: x ** 2)(t)

plt.figure(figsize=(8, 6))
plt.plot(t_pow2,t_pow3, label='Neilsche Parabel $y = t^3$, $x = t^2$')
plt.title('Neilsche Parabel')
plt.xlabel('$x = t^2$')
plt.ylabel('$y = t^3$')
plt.grid(True)
plt.legend()
plt.show()
```
